```python
import tensorflow_datasets as tfds
import os

from clearml import Dataset
tfds.__version__
```

```python
def make_clearml_ds(name, version=None, download=True):
    # name = 'oxford_flowers102'
    ds_name = name if version is None else f'{name}:{version}'
    print(f'==> processing dataset: {ds_name}')
    _, info = tfds.load(
        ds_name,
        split=['train','test'],
        as_supervised=True,
        with_info=True,
        download=download,
        # download_and_prepare_kwargs={'file_format': tfds.core.FileFormat.PARQUET}
    )
    
    print(info)
    ds_path = os.path.join('/root/tensorflow_datasets', name)
    if os.path.exists(ds_path) and info:
        ds = Dataset.create(dataset_name=name, dataset_project='tfds', dataset_version=str(info.version))
        ds.add_files(ds_path, dataset_path=name)
        print(ds.list_files())
        ds.finalize(auto_upload=True)
```

```python
for n in ['pet_finder']:
    if type(n)==str:
        make_clearml_ds(n)
    else:
        make_clearml_ds(*n)
```
